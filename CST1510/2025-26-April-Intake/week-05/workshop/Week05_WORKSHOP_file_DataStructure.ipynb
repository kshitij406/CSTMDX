## Files

- A **text file** consists of a series of characters.
- A **binary file** (used for images, videos, etc.) consists of a series of bytes.
- The first character in a text file, or byte in a binary file, is at position 0.
  - In a file with **_n_** characters or bytes, the last position is **_n_ – 1**.
- When you **open** a file in Python, it creates a **file object** that allows you to interact with the file.

### 2. The `with` Statement (Writing to a Text File)
- Python’s file handling capabilities are built on the [**`io` module**](https://docs.python.org/3/library/io.html), which provides a range of classes for working with files.

- The `with` statement:
  - Acquires a resource and assigns its corresponding object to a variable.
  - Enables the application to use the resource through that variable.
  - Automatically calls the resource object’s **`close` method** to release the resource when done.


````python
# Example:
with open('accounts.txt', mode='w') as f:
    f.write('101 Smith 150.75\n')
    f.write('202 Johnson 89.34\n')
    f.write('303 Williams 452.00\n')
    f.write('404 Brown -23.45\n')
    f.write('505 Davis 320.89\n')
````

### 4. Writing to a File
- `write`: Writes a string to the file.
- `writelines`: Takes an iterable and writes each element to the file.

````python
with open('accounts.txt', mode='w') as f:
    f.writelines(['101 Smith 150.75\n', '202 Johnson 89.34\n'])
````


### File Object Methods

### 6. File Modes

| Mode   | Description |
| ------ | ----------- |
| `'r'`  | Opens a text file for reading. This is the default mode if none is specified when calling `open`. |
| `'w'`  | Opens a text file for writing. Any existing content is **erased**. |
| `'a'`  | Opens a text file for appending. If the file doesn’t exist, it is created. New data is added to the end of the file. |
| `'r+'` | Opens a text file for both reading and writing. |
| `'w+'` | Opens a text file for both reading and writing. Any existing content is **erased**. |
| `'a+'` | Opens a text file for both reading and appending. If the file doesn’t exist, it is created. New data is added to the end of the file. |

## 1. `Lists`
Lists are mutable sequences that can store elements of different data types. They are versatile and provide various methods for manipulation.


### Common List Operations

````python
# Creating a list
numbers = [1, 3, 4, 5]
````

````python
# Adding an element
numbers.append(7)
print("After append:", numbers)  # Output: [1, 3, 4, 5, 7]
````

````python
# Accessing an element
print("First element:", numbers[0])  # Output: 1
````

````python
# Deleting an element
numbers.pop()  # Removes the last element
print("After pop:", numbers)  # Output: [1, 3, 4, 5]
````

````python
# Reversing the list
numbers.reverse()
print("Reversed list:", numbers)  # Output: [5, 4, 3, 1]
````

````python
# Summing elements in the list
print("Sum of elements:", sum(numbers))  # Output: 13
````

````python
# Clearing the entire list
numbers.clear()
print("After clear:", numbers)  # Output: []
````

## 3. `Sets`

- Sets are a collection of **distinct (unique) objects**.
- Sets behave as collections of dictionary keys without any
associated values
- These are useful to create lists that only hold unique values in the dataset.
- It is an unordered but mutable collection; this is very helpful when going
through a vast dataset.

### Common Set Operations

````python
# Creating a set
x = {3, 5, 3, 3}
print("Set with unique elements:", x)  # Output: {3, 5}
````

````python
# Converting a list to a set to remove duplicates
list1 = [1, 2, 1, 4]
x = set(list1)
print("Set from list:", x)  # Output: {1, 2, 4}
````

````python
# Basic set operations
x = {18, 21, 24, 27}
````

````python
# Adding an item
x.add(30)
print("After add:", x)  # Output: {18, 21, 24, 27, 30}
````

````python
# Removing an item
x.remove(18)
print("After remove:", x)  # Output: {21, 24, 27, 30}
````

````python
# Check membership
print("21 in set:", 21 in x)  # Output: True
````

````python
# Remove last element
x.pop()
print(x)
````

### List of functions on sets:

````python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
````

### Intersection ( `AND` )

````python
print(set1 & set2 )
````

````python
print(set1.intersection(set2))
````

### Union ( `OR` )

````python
print(set1 | set2)
````

````python
print(set1.union(set2))
````

### Symmetric Difference ( `XOR` )

````python
print( set1 ^ set2)
````

````python
print( set1.symmetric_difference(set2))
````

### Difference (In set1 but not in set2)

````python
print(set1 - set2)
````

````python
print(set1.difference(set2))
````

````python
print(set1)
set1.discard(3)
print(set1)
````

### Frozen set
The frozenset() function returns an unchangeable frozenset object (which is like a
set object, only unchangeable )

````python
mylist = [18, 21, 24, 27, 30]
x= frozenset(mylist)
print(x)
````

## 4. `Dictionaries`
Dictionaries store data as key-value pairs, making them ideal for associating related information.

### Common Dictionary Operations

````python
# Creating a dictionary with a single key-value pair
x = {'tomatoes': 15.5, 'onions': 3.50, 'mushrooms': 1.50}
````

````python
# Creating a dictionary using the `dict()` function with a list of tuples
x = dict([('tomatoes', 15.5), ('onions', 3.50), ('mushrooms', 1.50)])
````

````python
# Creating a dictionary using the `dict()` function with keyword arguments
x = dict(tomatos= 15.5, onions= 3.50, mushrooms= 1.50)
````

````python
# Adding or updating an item
x['tomatoes'] = 25.2
print("Updated dictionary:", x)  # Output: {'tomatoes': 25.2, 'onions': 3.5, 'mushrooms': 1.5}
````

````python
# Removing an item
del x['onions']
print("After delete:", x)  # Output: {'tomatoes': 25.2, 'mushrooms': 1.5}
````

````python
# Remove the last item from dict
x.popitem()
print(x)
````

````python
# Check membership in x (only looks in keys, not values)
'mushrooms' in x
````

````python
'mushrooms' not in x
````

### Accessing elements in a dictionary

````python
x = {'tomatoes': 15.5, 'onions': 3.50, 'mushrooms': 1.50}
````

````python
# returns list of keys in x?
x.keys()
````

````python
# returns list of values in x?
x.values()
````

````python
# returns list of key-value tuple pairs in x?
x.items()
````

````python
# iterate key/values pair
for key,value in x.items():
    print(key, value)
````

````python
# Iterating over key-value pairs
for key, value in x.items():
    print(f"{key}: {value}")
    # Output: 
    # tomatoes: 25.2
    # mushrooms: 1.5
````

````python
# iterate over keys
for k in x:
    print(k)
````

````python
# iterate values 
for v in x.values():
    print(v)
````

### Creating Data Structures

### Series
A Series is a one-dimensional array-like object that can hold data of any type

````python
# Creating a Series
data = [10, 20, 30, 40]
index=['a', 'b', 'c', 'd']

series = pd.Series(data,index)
print(series)
````

### DataFrame
A DataFrame is a two-dimensional table-like data structure with labeled axes (rows and columns).

````python
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}


df = pd.DataFrame(data)
print(df)
````
