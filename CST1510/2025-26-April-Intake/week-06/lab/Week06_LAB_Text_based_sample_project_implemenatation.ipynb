This week, you will focus on understanding and implementing a text-based **Shopping System**. This activity will build on the **Shopping System Implementation** from the Workshop and help you prepare for your chosen coursework project.

---

#### Instructions

1. **Review the Shopping System**:
   - Study the **Shopping System** example provided in the workshop materials. Pay close attention to how it handles:
     - Loading product data from a file
     - Displaying available products
     - Managing a cart where users can add items
     - Checking out and updating inventory
   
2. **Select a Coursework Project**:
   - In your **`Assessments 2 - Course Work`** folder, you will find two projects:
     - **CW1_Smart Vending Machine.pdf**
     - **CW2_Crypto Investment Platform.pdf**
   - Choose one of these projects for your coursework implementation. Both projects share some similarities with the Shopping System and will follow a similar workflow.

3. **Create a Text-Based Implementation**:
   - Start creating a text-based implementation for your chosen project.
   - Implement essential functions for your selected project
   - Use text files (`.txt`) for any necessary data storage (similar to the inventory and transaction files in the Shopping System) to manage item stock, user actions, and transaction logs.

4. **Test Your Code**:
   - Test each part of your code to ensure functionality and handle cases like out-of-stock items or invalid inputs.
   - Verify that your program handles both typical and edge cases smoothly.

5. **Reflect**:
   - Consider the similarities and differences between the Shopping System and your chosen project.
   - Document any challenges and your solutions, as this will aid your understanding and help with your report.

This lab activity will prepare you for the next stages of development, such as incorporating Object-Oriented Programming, client-server communication, and a graphical interface into your project.
