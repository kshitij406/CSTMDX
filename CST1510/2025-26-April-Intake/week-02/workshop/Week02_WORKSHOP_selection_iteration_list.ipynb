# Algorithms

Any computing problem can be solved by executing a series of actions in a specific order.
An algorithm is a procedure for solving a problem in terms of
- the actions to execute and
- the order in which these actions are executed
    
Specifying the order in which statements (actions) execute in a program is called program control.

# Control Statements

- **Sequential Execution:** Statements execute in the order they are written.
- **Transfer of Control:** Python allows you to change the execution flow using control statements.
    ### Types of Control
  - **Sequential Execution:** Code runs step-by-step.
  - **Selection (Conditional) Statements:** Directs the program to execute different code based on conditions.
  - **Iteration (Looping) Statements:** Repeats code multiple times based on conditions.

### Selection Statements

- **Three Types of Selection Statements:**
  - **`if` Statement:** Executes an action if a condition is `True`; skips the action if `False`.
  - **`if…else` Statement:** Executes one action if a condition is `True`, and a different action if `False`.
  - **`if…elif…else` Statement:** Chooses one of many actions based on multiple conditions.

- **Grouping Actions:** Wherever a single action is allowed, multiple actions can be used.


### Iteration Statements

- **Two Types of Iteration Statements:**
  - **`while` Statement:** Repeats an action as long as a condition is `True`.
  - **`for` Statement:** Repeats an action for each item in a sequence.

### `and`

| Condition 1 | Condition 2 | Expression            | Result      |
|-------------|-------------|-----------------------|-------------|
| `True`      | `True`      | `True and True`       | `True`      |
| `True`      | `False`     | `True and False`      | `False`     |
| `False`     | `True`      | `False and True`      | `False`     |
| `False`     | `False`     | `False and False`     | `False`     |


* Can write statements like this using a concise conditional expression.
* The parentheses are not required, but they make it clear that the statement assigns the conditional expression’s value to `result`.

### `and` Example


````python
result = ('Passed' if grade >= 60 else 'Failed')
````

````python
# Python checks the first condition (False) and stops, because the whole expression will be False.
if 5 > 10 and print("This won't be printed"):
    print("This won't run")
````

Explanation: In an `and` expression, if the first condition is `False`, Python doesn’t check the rest because the whole expression will be `False`. This improves performance by avoiding unnecessary checks

## `or`



| Condition 1 | Condition 2 | Expression            | Result      |
|-------------|-------------|-----------------------|-------------|
| `True`      | `True`      | `True or True`        | `True`      |
| `True`      | `False`     | `True or False`       | `True`      |
| `False`     | `True`      | `False or True`       | `True`      |
| `False`     | `False`     | `False or False`      | `False`     |

### Short-Circuit Evaluation for Better Performance

- Python stops checking an `and` condition as soon as it finds `False`.
- It stops checking an `or` condition as soon as it finds `True`.
- For `and` conditions, place the most likely `False` check first.
- For `or` conditions, place the most likely `True` check first.


### `or` Example

````python
# Python checks the first condition (True) and stops, because the whole expression will be True.
if 10 > 5 or print("This won't be printed"):
    print("This will run")
````

Explanation: In an `or` expression, if the first condition is `True`, Python stops and doesn’t check the rest because the whole expression will be `True`.

### `Switch` Statement
> **Note:** Available only in Python 10 and above.

A switch statement allows you to execute one of several branches based on a comparison to a single value.

Example:


````python
day = "Monday"
match day:
    case "Monday":
        print("Office Early")
    case "Tuesday":
        print("Office")
    case "Wednesday":
        print("Work From Home")
    case "Thursday":
        print("Gym then Office")
    case "Friday":
        print("Office Half Day")
    case _:
        print("Sleep In")
````

## `for` Statement
* Repeat an action or several actions for each item in a sequence of items.
* A string is a sequence of individual characters.

````python
for char in 'cst1510':
    print(char, end=' ')
````

* When the `for` loop starts, Python assigns the first character of `'cst1510'` (i.e., `'c'`) to the **target variable** between `for` and `in`.
* After executing the loop body, Python assigns the next character (`'r'`) to the target variable and repeats the process.
* This continues until all items in the sequence have been processed.
* It's common to use the target variable in the loop body, but it’s not mandatory.

## Built-In `range` Function and Generators

* Creates an iterable representing a sequence of consecutive integers starting from 0 and ending before the specified value.


````python
for counter in range(10):
    print(counter, end=' ')
````

* Function `range`’s two-argument version produces a sequence of consecutive integers from its first argument’s value up to, but not including, the second argument’s value

````python
for number in range(5, 10):
    print(number, end=' ')
````

Function range’s three-argument version produces a sequence of integers from its first argument’s value up to, but not including, the second argument’s value, incrementing by the third argument’s value (the step)

````python
for number in range(0, 10, 2):
    print(number, end=' ')
````

If the third argument is negative, the sequence progresses from the first argument’s value down to, but not including the second argument’s value, decrementing by the third argument’s value

````python
for number in range(10, 0, -2):
    print(number, end=' ')
````

# `break` and `continue` Statements

* A `break` statement instantly terminates a `while` or `for` loop.

````python
for number in range(100):
    if number == 10:
        break
    print(number, end=' ')
````

* A `continue` statement skips the rest of the current loop iteration.
  * In a `while` loop, it rechecks the condition before continuing.
  * In a `for` loop, it moves to the next item in the sequence (if available).


````python
for number in range(10):
    if number == 5:
        continue
    print(number, end=' ')
````

### Relational/ Comparison Operators on lists
You can compare lists using the comparison operators (>, >=, <, <=, ==, and !=). For this to work, the two lists must contain the same type of elements. The comparison uses lexicographical ordering: the first two elements are compared, and if they differ this deter- mines the outcome of the comparison; if they are equal, the next two elements are compared, and so on, until either list is exhausted.

````python
print([1,2,3] == [1,2,3])
````

````python
print([3,33,3] > [1,2,3])
````

````python
print([3,33,3] < [1,2,3])
````

````python
print([3,33,3] != [1,2,3])
````
