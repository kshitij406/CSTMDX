## Problem-Solving & Pseudocode Guide

1. Understand the Problem:
    - Identify the **inputs**, expected **outputs**, and any **constraints**.
2. Break Down the Problem:
    - Divide the problem into smaller steps. Each step should focus on a single task, like getting input, performing calculations, or displaying results.
3. Write Pseudocode:
    - Describe each step in simple language, focusing on the logic rather than code syntax.

```python
#Example:
#1. Get the width and height from the user
#2. Calculate area (width * height)
#3. Calculate perimeter (2 * (width + height))
#4. Display area and perimeter
```

4. Convert to Code:
    - Translate your pseudocode into actual programming code, step-by-step.

```python
# Step 1: Get width and height from user
width = float(input("Enter the width of the rectangle: "))
height = float(input("Enter the height of the rectangle: "))

# Step 2: Calculate area
area = width * height

# Step 3: Calculate perimeter
perimeter = 2 * (width + height)

# Step 4: Display area and perimeter
print(f"The area of the rectangle is: {area}")
print(f"The perimeter of the rectangle is: {perimeter}")
```

As a result, finding the distance between two point on the surface of the Earth is more complicated than simply using the Pythagorean theorem.

The distance between these points following the surface of the Earth, in kilometres is:

Distance = 6371.01 x arccos(sin(t_1) x sin(t_2) + cos(t_1) x cos(t_2)x cos(g_(1 )- g_(2 )))

The value 6371.01 in the previous equation wasn’t selected at random. It is the average radius of the Earth in kilometres.
