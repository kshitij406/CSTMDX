### Printing Multiple Lines of Text with a Single Statement
* A backslash (`\`) in a string acts as an **escape character**.
* The backslash followed by another character forms an **escape sequence**.
* `\n` is the **newline character** escape sequence, instructing `print` to move the output cursor to the beginning of the next line.


```python
print('Welcome\nto\n\nPython!')
```

### The `sep` parameter in print()

The separator between the arguments to print() function in Python is space by default (softspace feature) , 
which can be made to any character, integer or string as per our choice. 

```python
#For formatting a date
print('25','09','2024', sep='/')
  
#Another example
print('a.wieclaw','mdx.ac.uk', sep='@')
```

### `end` parameter in print()

By default, Python‘s print() function ends with a newline. 
A programmer with C/C++ background may wonder how to print without a newline. 
Python’s print() function comes with an `end` parameter. 
By default, the value of this parameter is ‘\n’, i.e. the new line character.


```python
# Ends the output with a space
print("Welcome to", end=' ')
print("MDX")
```

| `\\` | Insert a backslash character in a string.
| `\"` | Insert a double quote character in a string.
| `\'` | Insert a single quote character in a string.

### Formatting float value


```python
pi = 3.14159
print('Pi to 2 decimal places: %.2f' % pi) #%.2f: Limits the float to 2 decimal places in C-style formatting
```

```python
print('Pi to 2 decimal places: {:.2f}'.format(pi)) #{:.2f}: Similarly limits the float to 2 decimal places using both
```

```python
print(f'Pi to 2 decimal places: {pi:.2f}') # .format() and f-strings.
```


The format specifier inside the curly braces follows the Python format string syntax https://docs.python.org/3/library/string.html#format-string-syntax . 

Specifically, in this case, it consists of the following parts:

The empty string before the colon means "take the next provided argument to format()" – in this case the x as the only argument.

- The **10.2f** part after the colon is the **format specification**.
- The **f** denotes **fixed-point notation**.
- The **10** is the **total width** of the field being printed, lefted-padded by spaces.
- The **2** is the **number of digits** after the decimal point.

| Function         | Description                                                                                     | Example                    |
| :--------------- | :---------------------------------------------------------------------------------------------- | :------------------------- |
| `abs(x)`         | Returns the absolute value of `x`.                                                              | `abs(-2)` is `2`           |
| `max(x1, x2, ...)` | Returns the largest among `x1, x2, ...`.                                                     | `max(1, 5, 2)` is `5`      |
| `min(x1, x2, ...)` | Returns the smallest among `x1, x2, ...`.                                                    | `min(1, 5, 2)` is `1`      |
| `pow(a, b)`      | Returns `a` raised to the power of `b`. Same as `a ** b`.                                       | `pow(2, 3)` is `8`         |
| `round(x)`       | Returns the integer nearest to `x`. If `x` is equally close to two integers, the even one is returned. | `round(5.4)` is `5`<br>`round(5.5)` is `6`<br>`round(4.5)` is `4` |
| `round(x, n)`    | Returns the float value rounded to `n` digits after the decimal point.                          | `round(5.466, 2)` is `5.47`<br>`round(5.463, 2)` is `5.46` |


#### Example:

```python
# 2. Maximum Value
print("Maximum value in the list:", max(2, 5, 3, 8, 10))
```

```python
# 3. Minimum Value
print("Minimum value in the list:", min(2, 5, 3, 8, 10))
```

```python
# 4. Power Calculation (e.g., 2 raised to the power of 3)
print("2 raised to the power of 3:", pow(2, 3))
```

```python
# 5. Rounding Numbers
print("Rounding 5.4:", round(5.4))
print("Rounding 5.5:", round(5.5))
print("Rounding 4.5:", round(4.5))
```

```python
# 6. Rounding with Decimal Places
print("Rounding 5.466 to 2 decimal places:", round(5.466, 2))
print("Rounding 5.463 to 2 decimal places:", round(5.463, 2))
```

#### Activity

Write a program calculating a distance between this two points.
Asking user for the coordinates, perform the calculation and print out
the result (using formatting ).

$d = \sqrt {\left( {x_1 - x_2 } \right)^2 + \left( {y_1 - y_2 } \right)^2 }$

```python
import math

(x1,y1) = float(input('Enter coordinates x,y for point a '))
(x2,y2) = float(input('Enter coordinates x,y for point b '))

distance = math.sqrt(math.pow((x1-x2),2) + math.pow((y1-y2),2))

print(f'The distance between point a = {x1},{y1} and b = {x2},{y2} is {round(distance,2)}')
```
